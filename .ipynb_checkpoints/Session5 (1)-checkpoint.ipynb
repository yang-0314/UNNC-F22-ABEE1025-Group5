```python
# This is a function that I can run in Jupyter Notebook
def most_frequent_number(int_list):
    res_dict = {} # key is the integer, the value is the appreance times, e.g. {4: 3} means
                    # 4 has appeared for 3 times
    current_most_frequency_number = None
    current_most_frequent_times = None
    #  [1, 4, 12, 53, 21, 11, 12, 21, 22, 2, 3, 4, 6, 11, 12, 4, 11]
    for this_item in int_list:
        print('=============this iteration=================')
        print(f'integer: {this_item}')
        if this_item in res_dict:    
            this_item_appreance_times = res_dict[this_item] # before
            this_item_appreance_times += 1
            res_dict[this_item] = this_item_appreance_times # after I met this integer again
            # test whether or not this item is the current most frequent number
            if current_most_frequency_number is None:
                current_most_frequency_number = this_item
                current_most_frequent_times = this_item_appreance_times
            else:
                if this_item_appreance_times > current_most_frequent_times:
                    # do result updating
                    current_most_frequent_times = this_item_appreance_times
                    current_most_frequency_number = this_item
                elif this_item_appreance_times == current_most_frequent_times:
                    # draw condition
                    if this_item <= current_most_frequency_number:
                        # do result updating
                        current_most_frequent_times = this_item_appreance_times
                        current_most_frequency_number = this_item
                    else:
                        pass
                else:
                    pass
        else:
            # this is the first time to encounter this item
            res_dict[this_item] = 1
            if current_most_frequency_number is None:
                current_most_frequency_number = this_item
                current_most_frequent_times = 1
            
        print(f'results after this iteration: {current_most_frequency_number}, {current_most_frequent_times}')
    return (current_most_frequency_number, current_most_frequent_times)
```

```python
int_list = [1, 4, 12, 53, 21, 12, 12, 21, 22, 2, 3, 4, 6, 11, 12, 4, 11]
most_frequent_number(int_list)
```

## OOP

Remember in our Lect2 of BIM, object oriented model has the following features:

- Object/Instance: the basic run-time units that represent actual entities in the program

- Class: defines a template of the objects that have similar attributes and behaviors 

- Attributes and methods: represents the properties and behavior of objects
 
- Inheritance: enables objects to reuse the common attributes and methods

- Polymorphism: enables operation to have different behaviors in different instances

- Data Encapsulation: hides the complexity and details for simple reuse

```python
class Cat():
    def __init__(self, color, breed, 
                 birthday, name, owner):
        self.color = color
        self.breed = breed
        self.birthday = birthday
        self._name = name
        self.owner = owner
    
    def get_sound(self):
        return 'Meow'
    
    def get_age(self, current_year):
        return current_year - int(self.birthday)
        
    # getter
    @property
    def name(self):
        return self._name
    
    # setter
    @name.setter
    def name(self, new_name):
        if (type(new_name) is str 
            and len(new_name) < 10
            and new_name != 'dog'):
            self._name = new_name
        else:
            print('Your name is illegal!')
    
 
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    def __init__(self, color, breed, 
                 birthday, name, owner):
        self.color = color
        self.breed = breed
        self.birthday = birthday
        self._name = name
        self.owner = owner + '2022'
    
    def get_age(self, current_year):
        return current_year - int(self.birthday)
    
    @abstractmethod
    def get_sound(self):
        return None
        
    # getter
    @property
    def name(self):
        return self._name
```

```python
# This is assigned to Peter
class Cat(Animal):
    
    def get_sound(self):
        return 'Meow'
```

```python
# This is assigned to Dezhou
class Dog(Animal):
    def get_sound(self):
        return 'Woof'
    
    def get_age(self, current_year):
        return 1000
```

#### A cooling simulation example



An office room has 10 people working in there.Each occupant emits 120W heat. 
The arrival time of people follows a normal distribution, with the mean 9:00AM and standard deviation 20min. Occupants will leave the office for lunch in the noon. The mean leave time for lunch is 12:00 noon with 10min standard deviation. Each occupant will spend a mean of 60min with 10min standard deviation for the lunch break. After the lunch break, occupant will return to the office. Occupants get off the work at 6:00PM (mean) with 20min standard deviation. 
During the work, after staying more than 60min in the office, each occupant has 40% probability to go outside for a short break. The break time lasts for a mean of 15min with standard deviation of 5min. After the break, occupant will go back to the office. 


The A/C of the room is operated from 7:00AM to 19:00PM. The A/C is controlled using an bang-bang controller with 0.5C deadband. 

Write a program to simulate indoor air temperature variations in this room. The simulation time step is 1min. 


```python
from cooling_sim import Occupant, Outdoor, AC, Room
import numpy as np
import matplotlib.pyplot as plt
```

```python

occupants = [Occupant(heat_emit=120, mean_arrival=9, std_arrival=20,
    			mean_lunch=12, std_lunch=10, mean_lunch_time=60, 
    			std_lunch_time=10, mean_off=18, std_off=20, out_thres=60,
    			out_prob=0.4, mean_break=15, std_break=5) for i in range(10)]
occupants_total = [[] for i in range(10)]
while occupants[0].sim_time < 1440:
    for o_i in range(10):
        this_o = occupants[o_i]
        occupants_total[o_i].append(this_o.is_in_office)
        this_o.simulate_one_step()

occupants_total = np.array(occupants_total)
occupants_sche = occupants_total.sum(axis = 0)
print(occupants_sche)
```

```python
plt.plot(occupants_sche)
```

```python
outdoor = Outdoor(file_path='outdoor_air_temperature.csv', time_col='time', 
                  time_zone='US/Eastern', target_col='c', date='2017-07-01')
oat_res = []
for i in range(1440):
    outdoor.simulate_one_step()
    oat_res.append(outdoor.current_oat)
```

```python
occupants = [Occupant(heat_emit=120, mean_arrival=9, std_arrival=20,
    			mean_lunch=12, std_lunch=10, mean_lunch_time=60, 
    			std_lunch_time=10, mean_off=18, std_off=20, out_thres=60,
    			out_prob=0.4, mean_break=15, std_break=5) for i in range(10)]
outdoor = Outdoor(file_path='outdoor_air_temperature.csv', time_col='time', 
                  time_zone='US/Eastern', target_col='c', date='2017-07-01')
t_in = 28
t_set = 22
room =  Room(C=0.000001, wall_u = 1.0, wall_a = 80, c_p=1200, t_in_init=t_in, t_ac=12, sim_dt = 60)
ac = AC(m=0.2, dead_band=0.5)
t_in_res = []
for i in range(1440):
    # Occupant simulation
    occp_num = 0
    for o_i in range(10):
        this_o = occupants[o_i]
        this_o.simulate_one_step()
        occp_num += this_o.is_in_office
    # Outdoor simulation
    outdoor.simulate_one_step()
    oat = outdoor.current_oat
    # AC simulation
    ac.simulate_one_step(t_in, t_set)
    m = ac.current_m
    # Room simulation
    room.simulate_one_step(occp_num, oat, m)
    t_in = room.t_in
    t_in_res.append(t_in)
    
```
